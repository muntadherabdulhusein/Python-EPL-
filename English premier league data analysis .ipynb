```python
#Great a new 2 columns 
epl_df['MinsPerMatch']= (epl_df['Mins']/epl_df['Matches']).astype(int)
epl_df['GoalPerMatch']= (epl_df['Goals']/epl_df['Matches']).astype(float)
```

```python
epl_df.head()
```

```python
#Total goals for the season 
Total_Goals = epl_df['Goals'].sum()
print(Total_Goals)
```

```python
Total_Penalty_Goals = epl_df['Penalty_Goals'].sum()
print(Total_Penalty_Goals)
```

```python
chelsea_penalty_data = epl_df[(epl_df['Club'] == 'Chelsea') & (epl_df['Penalty_Goals'] != 0)]
print(chelsea_penalty_data.sum())
```

```python
Total_Penalty_Attempted = epl_df ['Penalty_Attempted'].sum()
print(Total_Penalty_Attempted)
```

```python
plt.figure(figsize=(13, 6))
pl_not_scored= epl_df['Penalty_Attempted'].sum() - Total_Penalty_Goals
data = [pl_not_scored,Total_Penalty_Goals]
labels = ['Penalties missed', 'Penalties scored']
color = sns.color_palette('Set2')
plt.pie(data, labels = labels, colors = color, autopct = '%.0f%%')
plt.show()
```

```python
under_20 = epl_df[epl_df['Age'] <= 20]
age20_25 = epl_df[(epl_df['Age']> 20) & (epl_df['Age'] <= 25)]
age25_30 = epl_df[(epl_df['Age']> 25) & (epl_df['Age'] <= 30)]
above30 = epl_df[epl_df['Age']>30]
```

```python
x = np.array([under_20['Name'].count(),age20_25['Name'].count(), age25_30['Name'].count(), above30['Name'].count()])
MyLabels=["<=20", ">20 & <25", ">25 & <30", ">30"]
plt.title('Total Player With Age Groups', fontsize = 20)
plt.pie(x, labels= MyLabels, autopct='%.1f%%')
plt.show
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x = 'Club', y= 'Age', data=epl_df)
plt.xticks(rotation=90)
```

```python
num_player = epl_df.groupby('Club').size()
data = epl_df.groupby('Club')['Age'].sum()/num_player
data.sort_values(ascending=False)
```

```python
Assist_by_clubs= pd.DataFrame(epl_df.groupby('Club', as_index=False)['Assists'].sum())
sns.set_theme(style='whitegrid',color_codes=True)
ax=sns.barplot(x='Club',y='Assists',data=Assist_by_clubs.sort_values(by='Assists'), palette= 'Set2')
ax.set_xlabel("Club",fontsize=30)
ax.set_ylabel('Assists', fontsize=20)
plt.xticks(rotation=75,fontsize=15)
plt.rcParams["figure.figsize"]=(20,8)
plt.title('Plot Of Club vs Total Assists', fontsize=20)
```

```python
top_10_assists = epl_df[['Name','Club','Assists','Matches']].nlargest(n=10, columns ='Assists')
top_10_assists
```

```python
Goals_by_clubs=pd.DataFrame(epl_df.groupby('Club', as_index=False)['Goals'].sum())
sns.set_theme(style='whitegrid',color_codes=True)
ax=sns.barplot(x='Club',y='Goals',data=Goals_by_clubs.sort_values(by='Goals'), palette= 'rocket')
ax.set_xlabel("Club",fontsize=30)
ax.set_ylabel('Assists', fontsize=20)
plt.xticks(rotation=75,fontsize=15)
plt.rcParams["figure.figsize"]=(20,8)
plt.title('Plot Of Club vs Total Goals', fontsize=20)
```

```python
plt.figure(figsize=(14,7))
assists=epl_df['Assists'].sum()
data=[Total_Goals- assists, assists]
labels=['Goals without assists','Goals with assists' ]
color=sns.color_palette('Set1')
plt.pie(data,labels=labels,colors=color,autopct='%.0f%%')
plt.show()
```
